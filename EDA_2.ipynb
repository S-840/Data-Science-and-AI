# Exploratory Data Analysis:
Exploratory Data Analysis (EDA) is a method used to `analyse` and `summarize` datasets. In python, you can use the pandas library to perform EDA on a datasets.

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# import datasets
df = sns.load_dataset('titanic')
```

```python
df.drop(columns="deck", axis=1, inplace=True)  
# if you can run code again and again it may through because it work
#  is done when first time code is run 
```

```python
df.columns
```

```python
# visualize and explore
sns.countplot(data=df, x="survived")
```

```python
sns.barplot(data=df, x="survived", y="fare", hue="who")
```

```python
sns.countplot(x="pclass", data=df)
# Barplot only make categorical variables
```

```python
# data distribution
sns.histplot(data=df, x='age')
```

```python
df['age'].hist(bins=30, color='darkred', alpha=0.7)
# Histplot is made of numeric data (more unique values) 
```

```python
sns.boxenplot(data=df, y='age', x="survived", hue='who')           
# Box plot is made between categorical and numeric values
```

```python
# Scatterplot Both numeric values
sns.scatterplot(x='age', y='fare', data=df, hue="pclass")
plt.ylim(0,250)
```

```python
sns.pairplot(df,hue='survived')
```

```python
#stacked barchart/plot
df.groupby(['pclass', 'who'])['survived'].mean().unstack().plot(kind='bar', stacked=True)
```

```python
#swarm
sns.swarmplot(data=df, x="survived", y='fare', hue='who')
```

```python
sns.catplot(x='pclass', y='fare', hue='embarked', kind='point', data=df)
```

```python
# Relation of two numeric values
sns.lmplot(data=df, x='age', y='fare')
```

```python
# How to add column and make age group
# Grouping
df.columns
```

```python
df['age_groups'] = pd.cut(df['age'], [0, 10, 18, 30, 50, 80], 
    labels=['Kids', 'Child', 'Youth', 'Middle age', 'Old']) 
```

```python
df.head()
```

```python
df.columns
```

```python
df.groupby("age_groups")["survived"].mean().plot(kind="bar")
```

```python
sns.pairplot(df, hue='survived', vars=['age', 'fare', 'pclass'])
```

```python
df['family_group'] = pd.cut(df['sibsp'] + df['parch'] + 1, 
                            bins=[0, 1, 2, 20], 
                            labels=['Alone', 'Partner', 'Family'])
```
